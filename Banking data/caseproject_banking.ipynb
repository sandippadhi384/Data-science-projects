```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("banking_data.csv")
data
```

# -	What is the distribution of age among the clients?

```python
list1=data.age.unique()
list1
```

```python
no=[]
for i in list1:
  x=data["age"].loc[data["age"]==i].count()
  no.append(x)
no
```

```python
plt.scatter(list1,no,color="r")
plt.ylabel("Number of people")
plt.xlabel("Age")
plt.show()
```

# -	How does the job type vary among the clients?

```python
jobs=data["job"].unique()
job_no=[]
for i in jobs:
  x=data["job"].loc[data["job"]==i].count()
  job_no.append(x)
print(job_no)
print(jobs)
```

```python
data["job"].value_counts()
```

```python
plt.bar(jobs,job_no)
plt.xticks(fontsize=10,rotation=90)
for index,height in enumerate(job_no):
    plt.text(index,height,int(height),ha='center',va='bottom',size=8)
plt.ylabel("Number of people")
plt.xlabel("Job types")
plt.show()
```

# -	Are there any correlations between different attributes and the likelihood of subscribing to a term deposit

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["y"]=le.fit_transform(data["y"])
data
```

```python
a=data.select_dtypes(include=[np.number])
b=a.corr()
b
```

```python
import seaborn as sns
sns.heatmap(b, annot=True, cmap="coolwarm")
plt.title("correlation matrix")
plt.tight_layout()
plt.show()
```
